```python
import pandas as pd

train_df = pd.read_csv('/content/11785-TCR/dataset/pytdc/train_PyTDC.csv')
valid_df = pd.read_csv('/content/11785-TCR/dataset/pytdc/valid_PyTDC.csv')
test_df = pd.read_csv('/content/11785-TCR/dataset/pytdc/test_PyTDC.csv')

print(train_df)
print(valid_df)
print(test_df)
```

```python
!git clone https://github.com/s-kachroo/CMU-11785-TCR-CLEAN.git
```

```python
# Current:
# Index [0,1,2,...] == Entry
# epitope_aa == EC Number
# tcr == Sequence

# Future:
# tcr == Entry
# epitope_aa == EC Number
# tcr_full == Sequence

# Operation 1: Remove all rows with label 0
train_df_filtered = train_df[train_df['label'] == 1]
valid_df_filtered = valid_df[valid_df['label'] == 1]
test_df_filtered = test_df[test_df['label'] == 1]

# Operation 2: Keep only the required columns and rename them

# For train_df
train_df_filtered = train_df_filtered[['epitope_aa', 'tcr']].reset_index(drop=True)
train_df_filtered['Entry'] = train_df_filtered.index
train_df_filtered = train_df_filtered[['Entry', 'epitope_aa', 'tcr']]
train_df_filtered.rename(columns={'epitope_aa': 'EC Number', 'tcr': 'Sequence'}, inplace=True)

# For valid_df: start Entry from len(train_df_filtered)
valid_df_filtered = valid_df_filtered[['epitope_aa', 'tcr']].reset_index(drop=True)
valid_df_filtered['Entry'] = valid_df_filtered.index + len(train_df_filtered)
valid_df_filtered = valid_df_filtered[['Entry', 'epitope_aa', 'tcr']]
valid_df_filtered.rename(columns={'epitope_aa': 'EC Number', 'tcr': 'Sequence'}, inplace=True)

# For test_df: start Entry from len(train_df_filtered) + len(valid_df_filtered)
test_df_filtered = test_df_filtered[['epitope_aa', 'tcr']].reset_index(drop=True)
test_df_filtered['Entry'] = test_df_filtered.index + len(train_df_filtered) + len(valid_df_filtered)
test_df_filtered = test_df_filtered[['Entry', 'epitope_aa', 'tcr']]
test_df_filtered.rename(columns={'epitope_aa': 'EC Number', 'tcr': 'Sequence'}, inplace=True)

print(train_df_filtered)
print(valid_df_filtered)
print(test_df_filtered)
```

```python
import os
os.makedirs('/content/CMU-11785-TCR-CLEAN/app/data/' + run_name, exist_ok=True)
train_df_filtered.to_csv('/content/CMU-11785-TCR-CLEAN/app/data/' + run_name + '/filtered_train_df.csv', index=False)
valid_df_filtered.to_csv('/content/CMU-11785-TCR-CLEAN/app/data/' + run_name + '/filtered_valid_df.csv', index=False)
test_df_filtered.to_csv('/content/CMU-11785-TCR-CLEAN/app/data/' + run_name + '/filtered_test_df.csv', index=False)
```
